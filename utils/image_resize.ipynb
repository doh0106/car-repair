```python
from PIL import Image

base = '/content/drive/MyDrive/data/차량파손데이터/160. 차량파손 이미지 데이터/01.데이터/test' # resize 할 이미지 폴더
base_resize = '/content/drive/MyDrive/data/차량파손데이터/160. 차량파손 이미지 데이터/01.데이터/test_resize' # 저장 폴더


for filename in os.listdir(base):
    if filename.endswith('.jpg'):
        img_path = os.path.join(base, filename)
        img = Image.open(img_path)
        img_resized = img.resize((448, 448))
        new_path = os.path.join(base_resize, filename)
        img_resized.save(new_path)
```
